# Work With CSV

```python
from faker import Faker
import csv

output = open('data.CSV', 'w')
fake = Faker()

header = ['name', 'age', 'street', 'city', 'state', 'zip', 'lng', 'lat']
mywriter = csv.writer(output)
mywriter.writerow(header)

for r in range(1000):
    mywriter.writerow([
        fake.name(),
        fake.random_int(min=18, max=80, step=1),
        fake.street_address(),
        fake.city(),
        fake.state(),
        fake.zipcode(),
        fake.longitude(),
        fake.latitude()
    ])

output.close()
```

```python
with open('data.CSV') as f:
    myreader = csv.DictReader(f)
    headers = next(myreader)
    for row in myreader:
        print(row['name'])


    
```

# Json

```python
import json
output = open('data.json', 'w')  # ایجاد فایل JSON برای نوشتن
fake = Faker()  # ساخت شیء Faker برای تولید داده‌های تصادفی

alldata = {}  
alldata['records'] = []  # ایجاد کلید records با مقدار یک لیست خالی

for x in range(1000):
    data = {
        "name": fake.name(),
        "age": fake.random_int(min=18, max=80, step=1),
        "street": fake.street_address(),
        "city": fake.city(),
        "state": fake.state(),
        "zip": fake.zipcode(),
        "lng": float(fake.longitude()),
        "lat": float(fake.latitude())
    }
    alldata['records'].append(data)  # افزودن رکورد جدید به لیست records
```

```python
alldata
```

## Save Json 
+   Save alldata to -> data.json

```python

# For Persian
# ensure_ascii=False,
with open('data.json', 'w') as json_file:
    json.dump(alldata, json_file,ensure_ascii=False, indent=4)
    
```

## Read json

```python
with open('data.json', 'r') as json_file:
    # Step 3: Load the data
    data = json.load(json_file)
```
